# Add a measure table to the semantic model

```python
data = [(1, 'MeasureTable')]
columns = ['ID', 'Col1']
```

```python
measure_df = spark.createDataFrame(data, columns)
measure_df.show()
spark.sql("DROP TABLE IF EXISTS MeasureTable")
measure_df.write.format("delta").saveAsTable('MeasureTable')
```
