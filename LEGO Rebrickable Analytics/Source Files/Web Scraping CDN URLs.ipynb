# Web Scraping CDN URLs

```python
# Import modules
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Initialize variables
colors_url = ''
elements_url = ''
inventories_url = ''
inventories_parts_url = ''
inventories_sets_url = ''
parts_url = ''
parts_categories_url = ''
sets_url = ''
themes_url = ''
```

```python
# Make soup for web page
response = requests.get(
    'https://rebrickable.com/downloads/')
wp = response.text
soup = BeautifulSoup(wp, 'html.parser')
```

```python
# Define function for filtering links to just those with downloadable csv.gz files
def filterToDownloadFiles(link):
    if 'gz' in link.text:
        return True
    else:
        return False
```

```python
# Get all links and assign to the right variable
all_links = soup.find_all('a')
download_links = filter(filterToDownloadFiles, all_links)

for link in download_links:
    if ('colors' in link.text):
        colors_url = link['href'].split('/')[-1]
    elif ('elements' in link.text):
        elements_url = link['href'].split('/')[-1]
    elif ('inventories' in link.text):
        inventories_url = link['href'].split('/')[-1]
    elif ('inventory_parts' in link.text):
        inventory_parts_url = link['href'].split('/')[-1]
    elif ('inventory_sets' in link.text):
        inventory_sets_url = link['href'].split('/')[-1]
    elif ('parts' in link.text):
        parts_url = link['href'].split('/')[-1]
    elif ('part_categories' in link.text):
        parts_categories_url = link['href'].split('/')[-1]
    elif ('themes' in link.text):
        themes_url = link['href'].split('/')[-1]
    elif ('sets' in link.text):
        sets_url = link['href'].split('/')[-1]
```

```python
# Check variable outputs
print(colors_url)
print(elements_url)
print(inventories_url)
print(inventory_parts_url)
print(inventory_sets_url)
print(parts_url)
print(parts_categories_url)
print(sets_url)
print(themes_url)
```

```python
# Make URLs available for use in the pipeline
urls = [colors_url, elements_url, inventories_url, inventory_parts_url, inventory_sets_url, parts_url, parts_categories_url, sets_url, themes_url]
urls_string = '|'.join(urls)

mssparkutils.notebook.exit(urls_string)
```
